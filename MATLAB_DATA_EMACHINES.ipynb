```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
voltage=[100,110,120,130,140,150,160,170,180,190,200,210,220,230];
current=[8.85,9.825,7.418,8.834,7.363,10.89,11.34,13.71,2.527,2.748,2.274,2.465,1.258,0.9294];
df = pd.DataFrame({
    'Supply Voltage': voltage,
    'Current':current
});
df
```

```python
df.to_csv('volt_current.csv', index= False);
```

```python
from google.colab import files
files.download('volt_current.csv')
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['Supply Voltage'], df['Current'], '-o', color='green', linewidth=2, markersize=6)
plt.title('Supply Voltage vs Current Curve in No load test of a Single Phase Induction Motor')
plt.xlabel('Supply Voltage (V)')
plt.ylabel('Current (A)')
plt.grid(True)
plt.savefig('volt_current_curve.png', dpi=300, bbox_inches='tight')
plt.show()
```
